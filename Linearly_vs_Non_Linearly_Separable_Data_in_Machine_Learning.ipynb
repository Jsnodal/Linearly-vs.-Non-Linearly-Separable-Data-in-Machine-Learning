
 # **Understanding Linearly vs. Non-Linearly Separable Data in Machine Learning**



**One of the first steps in classification tasks is determining whether your data is linearly separable or non-linearly separable. Here’s a simple breakdown:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([[1,2], [2,3], [3,4], [5,6], [6,7], [7,8]])
y = np.array([0, 0, 0, 1, 1, 1])  # Two classes: 0 and 1

# Scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="bwr", edgecolors='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot of Data")
plt.show()
```

 **🔹 Linearly Separable Data**

**Data can be separated by a straight line (or hyperplane in higher dimensions).**

**Example: Classifying emails as spam or not, where a single feature (like word frequency) can separate them.**

**Models that work well: Logistic Regression, Support Vector Machine (SVM) with a linear kernel.**

**Linearly Separable Data**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Generate linearly separable data
X_linear, y_linear = make_blobs(n_samples=100, centers=2, n_features=2, random_state=42)

# Plot the data
plt.scatter(X_linear[:, 0], X_linear[:, 1], c=y_linear, cmap="bwr", edgecolors='k')
plt.title("Linearly Separable Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_linear, y_linear)
print("Logistic Regression Accuracy:", log_reg.score(X_linear, y_linear))
```

**The data above is lineary separable...the decison libraly wil be straigth line**

# Non Linear Boundary ,
**a single straight line cannot separate the data , a curved decison is needed**

**Non-Linearly Separable Data**

**A straight line cannot separate the classes.**

**Example: The classic "moons" dataset, where data points form curved shapes.**

**Solutions:**

**✅ Use SVM with RBF kernel (captures non-linearity).**

**✅ Transform features using Polynomial Features.**

**✅ Consider Neural Networks for complex decision boundaries.**

```python
from sklearn.datasets import make_moons

# Generate non-linearly separable data
X_nonlinear, y_nonlinear = make_moons(n_samples=100, noise=0.1, random_state=42)

# Plot the data
plt.scatter(X_nonlinear[:, 0], X_nonlinear[:, 1], c=y_nonlinear, cmap="bwr", edgecolors='k')
plt.title("Non-Linearly Separable Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_nonlinear, y_nonlinear)
print("Logistic Regression Accuracy:", log_reg.score(X_nonlinear, y_nonlinear))
```

**❌ Low accuracy (~50-70%) → Logistic Regression struggles.**
**❌ Linear decision boundary cannot fit this dataset.**

**✔ Linearly Separable Data → A straight line can separate classes.**

**✔ Non-Linearly Separable Data → You need curved boundaries (RBF SVM or Polynomial Features).**

**✔ Use scatter plots, logistic regression, SVM, and polynomial features to diagnose separability.**



# 🔹 How to Check?

1️⃣ Visualize your data using scatter plots.

2️⃣ Train Logistic Regression or Linear SVM – if accuracy is low, it might be non-linear.

3️⃣ Try Polynomial Features or RBF Kernel – if accuracy improves, your data is non-linear!

📌 Key Takeaway: If your data is linearly separable, keep it simple with Logistic Regression or Linear SVM. If not, explore feature transformations or non-linear models like SVM (RBF) or Neural Networks.
